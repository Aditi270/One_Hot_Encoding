```python
#ONE HOT ENCODING is converting the catergorical data into numerical data
#it creates a new column for every unique value in the original category colum where 0 means False and 1 means True 
#Lets see in this example how many colums we get after performing one hot encoding
pd.get_dummies(df, drop_first=True).shape
```

```python
#Higher number of features or columns can affect accuracy so we 
#come up with idea of Ensemble Selection which is nothing but limiting the 
#one hot encoding to the most frequent feeatures
```

```python
#lets find the top 10 frequent categories for the variable X2
df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
top_10 = [x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
#and now we make 10 binary variables 
for label in top_10:
    df[label] = np.where(df['X2']==label,1,0)

df[['X2']+top_10].head(40)
```

```python
#getting dummie variables for all the categorical vaariables
def one_hot_encode(df,variable, top_x_lables):
    for label in (top_x_lables):
        df[variable+ '_'+label] = np.where(df[variable]==label,1,0)
#read the data again
df = pd.read_csv('Merc_test.csv', usecols=['X1','X2','X3','X4','X5'])
#encode X2 into the most frequent categories
one_hot_encode(df,'X2',top_10)
df.head()
    
```
